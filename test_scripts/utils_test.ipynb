## TrueAfter Demo

Class returns true after specified number of calls

Used in MovingAverageFilterPlus to know when filter is warm

```python
# TrueAfter is True after specified number of step() calls
after_int = np.random.randint(0, 10)
print("Returns True After {} calls\nStarting...".format(after_int))

trueafter_test = TrueAfter(after_int)
for i in range(after_int + 2):
    trueafter_test.step()
    print(i+1, ": ", trueafter_test.isafter())
```

## MovingAverageFilterPlus Demo

Show usage of filter including intialization, startup phase (warm=False), and regular usage (warm=True)

cold_start=True starts with empty buffer (buffer is initialized to correct size but average calculations will reflect starting condition)

cold_start=False starts the buffer with all values set to initial_value

```python
initial_val = np.random.random() * 10
buffer_size = np.random.randint(0, 10)


mafp_test = MovingAverageFilterPlus(cold_start=True, initial_value=initial_val, size=buffer_size)
print("Initial Buffer: ", mafp_test.buffer, "\n")

new_vals = [5]*20

for iter, val in enumerate(new_vals):
    mafp_test.update(val)
    print("Iteration: ", iter + 1)
    print("Buffer Contents: ", mafp_test.buffer)
    print("Average: ", mafp_test.average())
    print("Trimmed Average: ", mafp_test.trimmed_average())
    print()
```
